```python
def resumetable(df):
    print(f'데이터 셋 형상 :{df.shape}')
    summary = pd.DataFrame(df.dtypes, columns=['Data Type'])
    summary['#NULL'] = (df == -1).sum().values # 피처별 -1 r개수
    summary['#unique'] = df.nunique().values
    summary['data cat'] = None
    for col in df.columns:
        if 'bin' in col or col =='target':
            summary.loc[col, 'data cat'] = '이진형'
        elif 'cat' in col:
            summary.loc[col, 'data cat'] = '명목형'
        elif df[col].dtype == np.int64:
            summary.loc[col, 'data cat'] = '순서형'
        elif df[col].dtype == np.float64:
            summary.loc[col, 'data cat'] = '연속형'
    return summary 
summary = resumetable(train)
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def write_percent(ax, total_size):
    for patch in ax.patches:
        width = patch.get_width()
        height = patch.get_height()
        left_coord = patch.get_x()
        percent = height/total_size * 100
        
        ax.text(x=left_coord+width/2,
                y=height + total_size*0.001,
                s=f'{percent:1.1f}%',
                ha='center'
               )
mpl.rc('font', size=15)
plt.figure(figsize=(7,6))

ax = sns.countplot(x='target', data=train)
write_percent(ax, len(train))
ax.set_title('target distribution')
```

##### 타깃값의 비율차이가 크면 비율이 작은 타깃값을 잘 예측하는 것이 중요
##### 각각의 고윳값마다 타깃값 비율이 다른 피처여야 모델링에 도움

```python
import matplotlib.gridspec as gridspec

def plot_target_ratio_by_feature(df, features, num_rows, num_cols, size=(12,18)):
    mpl.rc('font', size=9)
    plt.figure(figsize=size)
    grid = gridspec.GridSpec(num_rows, num_cols)
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    
    for i, feature in enumerate(features):
        ax = plt.subplot(grid[i])
        sns.barplot(x=feature, y='target', data=df, palette='Set2', ax=ax)    
```

```python
bin_features = summary[summary['data cat']=='이진형'].index
plot_target_ratio_by_feature(train, bin_features,6,3)
```

```python
nom_features = summary[summary['data cat']=='명목형'].index
plot_target_ratio_by_feature(train, nom_features,6,3)
```

```python
ord_features = summary[summary['data cat']=='순서형'].index
plot_target_ratio_by_feature(train, ord_features,6,3,(12,20))
# 결측값 자체에 타깃값 예측력이 있다면 고윳값으로 간주하여 결측값 처리 필요 없음
# 분석결과 ps_ind_14, ps_calc_04~ps_calc_14 피처 제거
# ps_ind_14 : 통계적 유효성 저하
# ps_calc_04~ps_calc_14 : 고윳값 별 타깃값 비율 차이 없음
```

```python
cont_features = summary[summary['data cat']== '연속형'].index
plt.figure(figsize=(12,16))
grid = gridspec.GridSpec(5,2)
plt.subplots_adjust(wspace=0.2, hspace=0.4)
for idx, cont_feature in enumerate(cont_features):
    train[cont_feature] = pd.cut(train[cont_feature], 5)
    
    ax = plt.subplot(grid[idx])
    sns.barplot(x=cont_feature, y='target', data=train, palette='Set2', ax=ax)
    ax.tick_params(axis='x', labelrotation=10)
# 분석 결과 : ps_calc_01 ~ ps_calc_03 피처 삭제 가능
```
